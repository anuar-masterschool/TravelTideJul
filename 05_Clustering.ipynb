# Clustering from the leftovers

```python
import pandas as pd

users = pd.read_csv("data/groups/leftovers.csv")
```

```python
users
```

# Preprocessing

```python
df_preprocessed = users.copy()
```

## Drop
- user_id
- Age Group
- home_city

```python
df_preprocessed.drop(["user_id", "Age Group", "home_city"], axis=1, inplace=True)
```

```python
df_preprocessed.info()
```

```python
df_preprocessed
```

## Encoding


```python
df_preprocessed["gender"].value_counts()
```

```python
df_preprocessed["gender"] = df_preprocessed["gender"].map({"O": 0, "M": 1, "F": 2})
```

```python
df_preprocessed
```

```python
df_preprocessed["home_country"].value_counts()
```

```python
df_preprocessed["is_usa"] = df_preprocessed["home_country"].map({"usa": 1, "canada": 0})
df_preprocessed.drop("home_country", axis=1, inplace=True)
```

```python
df_preprocessed
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Create scaler and fit to data
scaler = StandardScaler()
X = scaler.fit_transform(df_preprocessed)
```

```python
X.shape
```

```python
X
```

# Clustering

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
kmeans.labels_
```

# Understanding Clusters

```python
users["group"] = kmeans.labels_
```

```python
users
```

```python
import seaborn as sns

sns.scatterplot(data=users, x="Age", y="total_avg_revenue", hue="group")
```
