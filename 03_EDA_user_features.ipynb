# EDA for the created User Features


```python
import seaborn as sns
```

```python
import pandas as pd
user_features = pd.read_csv('data/user_features.csv')
```

```python
user_features
```

```python
user_features["total_revenue"] =  user_features["hotel_tot_rev"] + user_features[("flight_tot_rev")]
```

# Revenue Analysis

```python
user_features["trip_session_ratio"].value_counts()
```

```python
user_features["trip_session_ratio"].hist()
```

```python
user_features["total_revenue"].hist()
```

```python
sns.scatterplot(user_features,x="trip_session_ratio", y="total_revenue")
```

```python
sns.boxplot(user_features,x="trip_session_ratio", y="hotel_tot_rev")
```

```python
sns.boxplot(user_features,x="trip_session_ratio", y="flight_tot_rev")
```

# Age Groups

```python
age_order = ["Age Group 18-35", "Age Group 35-50", "Age Group 50-65", "Age Group 65+"]
```

```python
sns.barplot(user_features,x="Age Group", y="hotel_tot_rev", order=age_order)
```

```python
sns.barplot(user_features,x="Age Group", y="flight_tot_rev", order=age_order)
```

```python
sns.barplot(user_features,x="Age Group", y="hotels_no_bookings", order=age_order)
```

```python
sns.barplot(user_features,x="Age Group", y="flight_no_bookings", order=age_order)
```

```python
sns.barplot(user_features, x="Age Group", y="flight_avg_seats", order=age_order)
```

```python
sns.boxplot(user_features, x="Age", y="flight_avg_seats")
```

```python
sns.countplot(user_features,x="Age Group", order=age_order)
```

```python
sns.barplot(user_features, x="gender", y="total_revenue")
```

```python
sns.countplot(user_features, x="Age Group", order=age_order, hue="gender")
```
